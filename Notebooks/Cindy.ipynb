```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Data/MNCAATourneyDetailedResults.csv')
```

```python
df
```

```python
def possessions(row):
    if 'WFGA' in row:
        possessions = 0.96*(row.WFGA + row.WTO + 0.44*row.WFTA - row.WOR)
    else:
        possessions = 0.96*(row.LFGA + row.LTO + 0.44*row.LFTA - row.LOR)
    return possessions

df['wPoss'] = df[['WFGA', 'WTO', 'WFTA', 'WOR']].apply(possessions, axis=1)
df['lPoss'] = df[['LFGA', 'LTO', 'LFTA', 'LOR']].apply(possessions, axis=1)
```

```python
# total possession
df['Poss'] = (df['wPoss'] + df['lPoss']) / 2
```
