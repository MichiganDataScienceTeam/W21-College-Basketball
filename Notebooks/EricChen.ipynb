```python
import pandas as pd
```

```python
dRegSeasonResults = pd.read_csv("MRegularSeasonDetailedResults.csv")
dTourneyResults = pd.read_csv("MNCAATourneyDetailedResults.csv")
```

```python
dRegSeasonResults
```

```python
# cumulative PIE% of winning and losing teams
# regular season results
dRegSeasonResults['WPIE%'] = dRegSeasonResults.apply(lambda row: ((row.WScore + row.WFGM + row.WFTM - row.WFGA - row.WFTA 
                  + row.WDR + 0.5*row.WOR + row.WAst + row.WStl + 0.5*row.WBlk - row.WPF - row.WTO) / (row.WScore 
                  + row.LScore + row.WFGM + row.LFGM + row.WFTM + row.LFTM - row.WFGA - row.LFGA - row.WFTA - row.LFTA 
                  + row.WDR + row.LDR + 0.5*row.WOR + 0.5*row.LOR + row.WAst + row.LAst + row.WStl + row.LStl 
                  + 0.5*row.WBlk + 0.5*row.LBlk - row.WPF - row.LPF - row.WTO - row.LTO)), axis=1)
dRegSeasonResults['LPIE%'] = dRegSeasonResults.apply(lambda row: ((row.LScore + row.LFGM + row.LFTM - row.LFGA - row.LFTA 
                  + row.LDR + 0.5*row.LOR + row.LAst + row.LStl + 0.5*row.LBlk - row.LPF - row.LTO) / (row.WScore 
                  + row.LScore + row.WFGM + row.LFGM + row.WFTM + row.LFTM - row.WFGA - row.LFGA - row.WFTA - row.LFTA 
                  + row.WDR + row.LDR + 0.5*row.WOR + 0.5*row.LOR + row.WAst + row.LAst + row.WStl + row.LStl 
                  + 0.5*row.WBlk + 0.5*row.LBlk - row.WPF - row.LPF - row.WTO - row.LTO)), axis=1)

dRegSeasonResults
```
